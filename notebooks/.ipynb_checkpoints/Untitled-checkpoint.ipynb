```python
import smtplib
from email.message import EmailMessage
```

```python
def send_csv_email(file_path, subject, receiver):
    msg = EmailMessage()
    msg['Subject'] = subject
    msg['From'] = "gfernandes2108@gmail.com"
    msg['To'] = receiver
    msg.add_alternative("""
    <html>
      <body>
        <h1>Pipeline finished!</h1>
        <p>Everything ran successfully!</p>
      </body>
    </html>
    """, subtype='html')

    with open(file_path, 'rb') as f:
        msg.add_attachment(
            f.read(),
            maintype='text',
            subtype='csv',
            filename=file_path
        )
    with smtplib.SMTP('smtp.gmail.com', 587) as server:
        server.starttls()
        server.login("gfernandes2108@gmail.com","gxyr ywbs vrla piyp")
        server.send_message(msg)
    
    print("Email sent!")
```

```python
file_path = "../data/2026_data_challenge_ae_data.csv"
subject = "testing email"
receiver = "ghab12@hotmail.com"

send_csv_email(file_path, subject, receiver  )
```
